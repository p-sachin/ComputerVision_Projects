```python
# Importing Libraries
import  numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

4 Bitwise Operators (AND, OR, XOR, NOT) and is used to create masks to further process our images

```python
# First Create a black canvas with numpy with 600 x 600
black_canvas = np.zeros((600, 600), np.uint8)
plt.figure(figsize=(10,6))
plt.imshow(black_canvas, cmap='gray')
```

```python
# Make the copy of the canvas and draw a white square on it (240 x 240)
square = black_canvas.copy()
cv2.rectangle(square, (300, 300), (60, 60), (255,255,255), cv2.FILLED)
plt.figure(figsize=(10,6))
plt.imshow(square, cmap='gray')
```

```python
# Make the copy of the canvas and draw a white circle on it
circle = black_canvas.copy()
cv2.circle(circle, (290,290), 150, (255,255,255), cv2.FILLED)
plt.figure(figsize=(10,6))
plt.imshow(circle, cmap='gray')
```

## Bitwise Operators

```python
# Performing Bitwise Operations
# AND: A bitwise AND is true if and only if both pixels are greater than zero.
square_copy = square.copy()
circle_copy = circle.copy()
sqr_and_circ = cv2.bitwise_and(square_copy, circle_copy)
plt.figure(figsize=(10,6))
plt.imshow(sqr_and_circ, cmap='gray')
```

```python
# Performing Bitwise Operations
# OR: A bitwise OR is true if either of the two pixels is greater than zero.
square_copy = square.copy()
circle_copy = circle.copy()
sqr_or_circ = cv2.bitwise_or(square_copy, circle_copy)
plt.figure(figsize=(10,6))
plt.imshow(sqr_or_circ, cmap='gray')
```

```python
# Performing Bitwise Operations
# A bitwise XOR is true if and only if one of the two pixels is greater than zero, but not both. 
square_copy = square.copy()
circle_copy = circle.copy()
sqr_xor_c = cv2.bitwise_xor(square_copy, circle_copy)
plt.figure(figsize=(10,6))
plt.imshow(sqr_xor_c, cmap='gray')
```

```python
# Performing Bitwise Operations
# NOT Operation on two images
# NOT: A bitwise NOT inverts the “on” and “off” pixels in an image.
sqr_not = cv2.bitwise_not(square)
plt.figure(figsize=(10,6))
plt.imshow(sqr_not, cmap='gray')
```

```python
# Performing Bitwise Operations
# NOT Operation on two images
circle_not = cv2.bitwise_not(circle)
plt.figure(figsize=(10,6))
plt.imshow(circle_not, cmap='gray')
```
