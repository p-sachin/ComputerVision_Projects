```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from yolo_helpers import *
```

```python
def imgshow(img):
    plt.figure(figsize=(20,15))
    plt.imshow((img[:,:,::-1]))
```

```python
models_config_path = 'models/yolov3.cfg'
models_weights_path = 'models/yolov3.weights'
```

```python
net = cv2.dnn.readNetFromDarknet(models_config_path, models_weights_path)
```

```python
img = cv2.imread('../img/dogs.jpg')
imgshow(img)
```

```python
# Create a blob of the image
blob = cv2.dnn.blobFromImage(img, 1/255.0, (416,416), swapRB=True)

#Set the input of the network
net.setInput(blob)

#Get the output layers
#net.getLayerNames()

output_layers = getOutputsNames(net)

# Forward pass for predictions
prediction = net.forward(output_layers)

#Post Process
postprocess(img, prediction)

imgshow(img)
```
