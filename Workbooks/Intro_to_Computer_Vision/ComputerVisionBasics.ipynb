## Import OpenCV

```python
# Import OpenCV
import cv2
```

```python
# Check OpenCV Version
cv2.__version__
```

# Load the Image

```python
# Load the Image
# Declare a variable with the path to your image
path = '../img/'
my_img = cv2.imread(path + 'food.jpg')
```

## Check image type

```python
# Check image type
# What is the image type? -> type
type(my_img)
```

## Display Pixel Values

```python
# Print the values of the pixels of the image
print(my_img)
```

## Show the image using OpenCV

```python
# Lets now see the image! dont forget to call destroyAllWindows!
cv2.imshow('My Image', my_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## Display using Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize=(15,10))
plt.imshow(my_img)
```

## Fixing the Colors

```python
# Does your image look ok? are the colors the right ones?, if not, lets fix it!
# Remember that openCV will load images using BGR rather than RGB
# Change the image to RGB and diplay it again with plt
rgb_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.imshow(rgb_img)
```

## Image Shape

```python
# What is the shape of the image?
# Check the image shape
rgb_img.shape
```

## Convert to GrayScale

```python
# Lets make it grayscale
# Load the the image in grayscale, what are 2 ways you can do that?
food_grayscale= cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
cv2.imshow('GrayScale Food', food_grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## Cropping an image

```python
# Cropping an image
# Challenge, you know an image is nothing else than a NumPy ndarray, knowing that, how could you
# Crop an image?
# Lets now crop a Region of Interest of an image, load and image and crop different parts of it
# To crop an image, the format to do so is, image[start_row:end_row, start_column:end_column]

food_cropped = rgb_img[50:600, 50:900]
cv2.imshow('Cropped Image', food_cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Save to disk the cropped areas
cv2.imwrite(path + 'food_cropped.jpg', food_cropped)
```

## Annotate the images

```python
# Lets now annotate some images
# To annotate an image making a copy of it is always a good idea!
# Load an image and then make a copy of it
img_cat = cv2.imread(path + 'cat.jpg')
cat_copy = img_cat.copy()
```

```python
cv2.imshow('Cat', cat_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## Rectangles

```python
# Annotating images
# Rectangles
# Draw green rectangle on top of a region of interest in your image
rgb_cat = cv2.cvtColor(cat_copy, cv2.COLOR_BGR2RGB)
# plt.figure(figsize=(15,10))
# plt.imshow(rgb_cat)
cv2.rectangle(rgb_cat, (280, 597), (650, 10), (0,255,0), 3) # 

plt.figure(figsize=(15,10))
plt.imshow(rgb_cat)
```

## Adding Text to Images

```python
# Make a copy of your image and then write some text on top of the image

cat_rgb_text = rgb_cat.copy()
cv2.putText(cat_rgb_text, 'Hello Cat!!', (320,30), cv2.FONT_HERSHEY_PLAIN, 3, (0,255,0), 3)
plt.figure(figsize=(15,10))
plt.imshow(cat_rgb_text)
```
