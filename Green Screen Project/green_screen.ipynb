```python
# Import the Libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Display HSV ColorMap as a helper for selecting color pixels
colormap = cv2.imread('img/hsv.png')
colormap_rgb = cv2.cvtColor(colormap, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.imshow(colormap_rgb)
```

```python
# Reading the images
man = cv2.imread('img/green-man.jpg')
moon = cv2.imread('img/moon-surface.jpg')

# Converting the images from BGR to RGB
man_rgb = cv2.cvtColor(man, cv2.COLOR_BGR2RGB)
moon_rgb = cv2.cvtColor(moon, cv2.COLOR_BGR2RGB)

# Convert the image from BGR to HSV
man_hsv = cv2.cvtColor(man, cv2.COLOR_BGR2HSV)
moon_hsv = cv2.cvtColor(moon, cv2.COLOR_BGR2HSV)

hsv_man = man_hsv.copy()
# Selecting the Upper and Lower Bounds for Green 
lower_bounds = (40, 50, 20)
upper_bounds = (80, 255, 255)
# Select mask region 
mask = cv2.inRange(hsv_man, lower_bounds, upper_bounds)
masked_man = man.copy()
masked_moon = moon.copy()
# Applying mask 
masked_man[mask != 0] = [0,0,0]
masked_moon[mask == 0] = [0,0,0]
merged = masked_man + masked_moon
final_image = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)


#Displaying all the images in subplots
fig, ax = plt.subplots(3, 1, figsize=(25,15))
ax[0].imshow(man_rgb)
ax[0].set_title("Original Image (Man)")
ax[1].imshow(moon_rgb)
ax[1].set_title("Original Image (Moon Surface)")
ax[2].imshow(final_image)
ax[2].set_title("Final Image")

```
