```python
# Importing  the Libraries
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np
```

```python
# Reading the Image
dog = cv2.imread('img/dog.jpg')
dog_rgb = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)

# Create a Dictionary the objects bbox and text fills boundaries
object_dict = {
        'dog': [[(130, 540), (310, 220)], [(200, 220), (130, 195)]],
        'bicycle': [[(125, 420), (565, 135)], [(125, 135), (210,110)]],
        'truck': [[(460, 170), (700, 75)], [(460, 75), (560, 50)]]
}

# Selecting HERSHEY DUPLEX Font Style
font_style = cv2.FONT_HERSHEY_DUPLEX

# Create a Copy of the original image
copy_dog = dog_rgb.copy()

# Creating the rectange bbox on the objects with text annotations
# For Dog
cv2.rectangle(copy_dog, object_dict['dog'][0][0], object_dict['dog'][0][1], (255,20,147), 2)
cv2.rectangle(copy_dog, object_dict['dog'][1][0], object_dict['dog'][1][1], (255,20,147), cv2.FILLED)
cv2.putText(copy_dog, 'Dog', (150, 210), font_style, 0.5, (0,0,0), 1)

# For Bicycle
cv2.rectangle(copy_dog, object_dict['bicycle'][0][0], object_dict['bicycle'][0][1], (0,0,128), 2)
cv2.rectangle(copy_dog, object_dict['bicycle'][1][0], object_dict['bicycle'][1][1], (0,0,128), cv2.FILLED)
cv2.putText(copy_dog, 'Bicycle', (140, 125), font_style, 0.5, (255,255,255), 1)

# For Truck
cv2.rectangle(copy_dog, object_dict['truck'][0][0], object_dict['truck'][0][1], (255,215,0), 2)
cv2.rectangle(copy_dog, object_dict['truck'][1][0], object_dict['truck'][1][1], (255,215,0), cv2.FILLED)
cv2.putText(copy_dog, 'Truck', (490, 65), font_style, 0.5, (0,0,0), 1)

# Displaying the results using subplots

fig, ax = plt.subplots(1,2, figsize=(20,15))
ax[0].imshow(dog_rgb)
ax[0].set_title("Original Image")

ax[1].imshow(copy_dog)
ax[1].set_title("Final Image with Annotation")

```
