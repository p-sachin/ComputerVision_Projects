```python
# Importing the Libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the QR Code Image
qr_img = cv2.imread('img/qr-code.png')
qr_img = cv2.resize(qr_img, (600, 600))

# Create an object for the class detector for the QR code
detector = cv2.QRCodeDetector()

# Get the Data and the Bounding box 
data, bbox, qr_code = detector.detectAndDecode(qr_img)

# If the bounding box has been detected we sucessfully decoded the QR Code
if bbox is not None:
    # Making a subplot of all the images
    fig, ax = plt.subplots(1,3, figsize=(15,10))
    # Changing the shape of the bbox 
    bbox = bbox.reshape(-1,2)
    num_lines = len(bbox)
    # Displaying Original QR Code
    ax[0].imshow(qr_img)
    ax[0].set_title("Original QR Code")
    ax[0].set_xlabel("Width")
    ax[0].set_ylabel("Height")
    # Creating 4 boundary lines on the QR Code
    for index in range(num_lines):
        pt1 = tuple(bbox[index])
        pt2 = tuple(bbox[(index+1) % num_lines])
        cv2.line(qr_img, pt1, pt2, (0, 255, 0), 4)

# Displaying the result
    ax[1].imshow(qr_img)
    ax[1].set_title('QR Code Detected')
    ax[1].set_xlabel("Width")
    ax[1].set_ylabel("Height")

# Final QR Code
    ax[2].imshow(qr_code, cmap='gray')
    ax[2].set_title("Final QR Code")
    ax[2].set_xlabel("Width")
    ax[2].set_ylabel("Height")

# Printing the Decoded Data
    print(f'QR Code Detected with the data: {data}')

# Saving the Final QR Code
    final_qr_code = cv2.imwrite('img/qr-code-saved.jpg', qr_code)
```
